```python
import pandas as pd
from plot_helpers import plot_save_duration, plot_cpu_utilization, plot_percent_change, plot_throughput, plot_key_bytes_throughput
```

```python

num_keys = 50.0
value_size = 50
title = f"{int(num_keys)}M Keys, {value_size} Byte Values - "


no_compression_df = pd.read_csv(f"save_summary_{num_keys}keys_{value_size}B_comp-no_csum-yes.csv")
yes_compression_df = pd.read_csv(f"save_summary_{num_keys}keys_{value_size}B_comp-yes_csum-yes.csv")
# yes_compression_df = pd.read_csv(f"new_prefetch_save_summary_{num_keys}keys_{value_size}B_comp-no_csum-yes.csv")
 

plot_save_duration(yes_compression_df, no_compression_df)
plot_cpu_utilization(yes_compression_df, no_compression_df)
plot_percent_change(
    df_yes_compression=yes_compression_df.copy(), # Use .copy() to avoid modifying the original df
    df_no_compression=no_compression_df.copy(),
    metric_column='save_duration_seconds',
    title_prefix=title
)
plot_throughput(yes_compression_df, no_compression_df)
plot_key_bytes_throughput(yes_compression_df, no_compression_df)
```
